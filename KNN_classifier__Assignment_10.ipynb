***K-Nearest Neighbor (KNN) classifier***

From the dataset of USA Forensic Science Service which has description of 6 types of glass; defined in terms of their
oxide content (i.e. Na, Fe, K, etc). 
using K-Nearest Neighbor (KNN) classifier to classify the glasses.

```python
#importing the packages 
import pandas as pd 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report 
from scipy.spatial import distance
```

**Reading the Datasets using Pandas**

```python
headerNames = ['id', 'RI', 'Na','Mg','Al','Si','K','Ca','Ba','Fe','class attribute']  # creating a Names for Attributes
#Train dataset
train_file = pd.read_csv('/content/trainKNN.txt')  # Reading a text file as csv file
train_file.to_csv ('trainKNN.csv', header=headerNames, index=None)  # adding a Names to attributes of the file
```

```python
df1 = pd.read_csv('/content/trainKNN.csv')  
```

```python
df1 = df1.drop('id',axis=1)  # removing the column 'id'
```

```python
df1.head()
```

```python
df1.describe()
```

```python
#test dataset
test_file = pd.read_csv('/content/testKNN.txt')  # Reading a text file as csv file
test_file.to_csv('testKNN.csv', header=headerNames, index=None)  # adding a Names to attributes of the file
```

```python
df2 = pd.read_csv('/content/testKNN.csv')
```

```python
df2 = df2.drop('id', axis=1)  # removing the column 'id'
```

```python
df2.head()
```

```python
df2.describe()
```

**standardization**

```python
def standardize (df):      #function for Standardization
  for col in df.columns:
    if col != 'class attribute':
      df[col] = (df[col] - df[col].mean()) / df[col].std()
  return df 
```

```python
train = standardize(df1)    #standardize data
test = standardize(df2)     #standardize data
```

```python
train.head()
```

Model Training

```python
# k = 8 neighbors
euclid_model = KNeighborsClassifier(n_neighbors=8, metric=distance.sqeuclidean) # Square Euclidean distance model
manhattan_model = KNeighborsClassifier(n_neighbors=8, metric=distance.cityblock) # Manhattan distance model
```

```python
x_train = train.drop(["class attribute"], axis=1)
y_train = train["class attribute"]

x_test = test.drop("class attribute", axis=1) 
y_test = test["class attribute"]
# Train models
euclid_model.fit(x_train,y_train) 
manhattan_model.fit(x_train, y_train)
```

```python
# Make predictions
manhattan_predictions = manhattan_model.predict(x_test)
euclid_predictions = euclid_model.predict(x_test) 
df = pd.DataFrame({'actual': y_test, 'manhattan': manhattan_predictions, 'euclid': euclid_predictions})
df.head()
```

Evaluate Model 

```python
manhattan_count = len(df.loc[df['manhattan'] == df['actual']])
euclid_count = len(df.loc[df['euclid'] == df['actual']])
```

```python
print('Manhattan Accuracy: {}%'.format(round(100*manhattan_count/len(df), 2)))
print(classification_report(y_test, manhattan_predictions, target_names=df['actual'].astype(str).unique()))
print ('\n')
print('Square Euclidean Accuracy: {}%'.format(round(100*euclid_count/len(df), 2)))
print(classification_report(y_test, euclid_predictions, target_names=df['actual'].astype(str).unique()))
```
